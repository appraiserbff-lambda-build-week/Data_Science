```python
df_for_poly_reg = df_no_na_int[['latitude', 'longitude', 'regionidcounty',
       'fips', 'propertylandusetypeid', 'assessmentyear', 'bedroomcnt',
       'bathroomcnt', 'roomcnt', 'regionidzip', 'taxamount',
       'landtaxvaluedollarcnt', 'regionidcity', 'yearbuilt',
        'calculatedfinishedsquarefeet',
        'appraised_value_y']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


X = df_for_poly_reg.drop(['appraised_value_y', 'ln_price'], axis=1)
y= df_for_poly_reg['ln_price']

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=.25, random_state=42, shuffle=True)

model = LinearRegression()
model.fit(X_train, y_train)

y_test_predict = model.predict(X_test)

MSE = mean_squared_error(y_test, y_test_predict)

RMSE = (np.sqrt(mean_squared_error(y_test, y_test_predict)))

print('MSE is {}'.format(MSE))
print('RMSE is {}'.format(RMSE))

R2 = r2_score(y_test, y_test_predict)

print('R^2 is {}'.format(R2))

print("Coefficient: \n", model.coef_)
print("\n Intercept: ", model.intercept_)
```
